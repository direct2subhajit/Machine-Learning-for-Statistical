```python
import pandas as pd
import numpy as np
df=pd.read_csv("API_PA.NUS.FCRF_DS2_en_csv_v2_1679468.csv",skiprows=4)
df.head()
```

```python
df.shape
```

```python
#droping columns that are not needed
df = df.drop(['Unnamed: 65','Indicator Name','Indicator Code'], axis=1)
```

```python
df.shape
```

```python
#filling the columns having NaN with the previous column value
df.bfill(axis=1)
```

```python
#making a new dataframe containing required rows
df1 = df[(df['Country Name'].isin(['China','India','Great Britain','Canada','Japan','Switzerland']))]
df1
 
```

```python
#re-indexing
df1.reset_index(drop=True)
```

```python
df1.to_csv(r'cleaned_dataset.csv')
```
