```python
# drop the unrequired columns
gdf = gdf.drop(['SUBJECT','MEASURE','TIME', 'Unit Code','PowerCode Code','Reference Period Code',
                'Reference Period', 'Flag Codes','Flags'], axis=1)
```

```python
#making a new dataframe containing required rows
gdf1 = gdf[(gdf['Country'].isin(['China','India','United Kingdom','Canada','Japan','Switzerland']))]
gdf1
 
```

```python
gdf1=gdf1.rename(columns={'Country': 'Country Name', 'LOCATION':'Country Code', 'Time': 'Year' })
gdf1.head()
```

```python
gdf1['Unit'].fillna('Percentage', inplace=True)
gdf1.head(20)
```
